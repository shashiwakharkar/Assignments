```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
comp_df = pd.read_csv('Computer_Data.csv')
comp_df
```

```python
#comp_df.drop(['cd', 'multi', 'premium'], axis = 1)
comp_df.drop(comp_df.columns[[0, 6, 7, 8]], axis = 1, inplace = True)
comp_df
```

```python
comp_df.boxplot(grid = False)
```

```python
for column in comp_df:
    plt.figure()
    comp_df.boxplot([column])
```

```python
fig, ax = plt.subplots(figsize = (15, 12),)
comp_df.hist(ax = ax)
```

```python
sns.distplot(a = comp_df["price"], kde = True,
kde_kws={"color" : "g", "alpha" : 0.3, "linewidth" : 3, "shade" : False})
```
